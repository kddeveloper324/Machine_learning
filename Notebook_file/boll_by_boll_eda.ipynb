# Data Preprocessing & Feature Engineering
To make the data suitable for a machine learning model, we transformed the raw ball-by-ball data into improved numerical features:

        Chronological Sorting:
 The data was sorted by Season, Match, and Over to ensure the timeline was respected.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. Sort values to ensure chronological order for cumulative features
df = df.sort_values(['season_id', 'match_id', 'innings', 'over_number', 'ball_number'])
```

# Cumulative Features: 
We created "Contextual Features" that tell the model the state of the game before every ball:

innings_runs_so_far: The total score of the team at that moment.

innings_wickets_so_far: How many wickets had been lost previously.

total_balls_bowled: The exact ball count in the inning (e.g., the 34th ball of the innings).

```python

# 2. Feature Engineering
# Cumulative runs and wickets per innings
df['innings_runs_so_far'] = df.groupby(['match_id', 'innings'])['total_runs'].cumsum() - df['total_runs'] # Exclude current ball
df['innings_wickets_so_far'] = df.groupby(['match_id', 'innings'])['is_wicket'].cumsum() - df['is_wicket'].astype(int) # Exclude current ball

# Current Ball Number in innings
df['total_balls_bowled'] = df.groupby(['match_id', 'innings']).cumcount() + 1
```

#  Encoding Categorical Data
Machine learning models require numerical input. We applied Label Encoding to convert text data into unique integers:

Encoded Columns: batter, bowler, non_striker, team_batting, team_bowling, batsman_type, bowler_type.

Example: "Mumbai Indians" might become 5 and "Chennai Super Kings" might become 2.

```python

# 3. Encoding Categorical Data
le = LabelEncoder()
categorical_cols = ['batter', 'bowler', 'non_striker', 'team_batting', 'team_bowling', 'batsman_type', 'bowler_type']

mapping_dict = {}
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
    mapping_dict[col] = dict(zip(le.classes_, le.transform(le.classes_)))
```

```python

# 4. Prepare X and y
# Features that are known BEFORE the ball is bowled
features = ['innings', 'over_number', 'ball_number', 'batter', 'bowler', 'non_striker', 
            'team_batting', 'team_bowling', 'batsman_type', 'bowler_type', 
            'innings_runs_so_far', 'innings_wickets_so_far', 'total_balls_bowled']

target = 'total_runs'
```

# Train-Test Split (Time-Based)
Instead of splitting the data randomly (which would leak future trends into the past), we used a Time-Based Split:

Training Data: All matches before the 2024 season.

Testing Data: Matches from the 2024 season onwards. This tests the model's ability to predict "future" games.

```python

# 5. Split Data (Temporal Split)
# Use data before 2024 for training, 2024 and beyond for testing
train_df = df[df['season_id'] < 2024]
test_df = df[df['season_id'] >= 2024]

X_train = train_df[features]
y_train = train_df[target]
X_test = test_df[features]
y_test = test_df[target]

print(f"Training Shape: {X_train.shape}")
print(f"Testing Shape: {X_test.shape}")
```

# Model Training
We trained a Random Forest Classifier to predict the outcome of the next ball (specifically, the total_runs scored: 0, 1, 2, 3, 4, 6, etc.).

```python
# Initialize and train the model
# Using a smaller number of estimators and depth for speed in this demo environment
rf_model = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42, n_jobs=-1)

rf_model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = rf_model.predict(X_test)
```

```python
# Evaluation
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, zero_division=0)
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", report)
```

# Model Performance & Results
1. Prediction Accuracy: ~45%

The model correctly predicted the exact runs scored on the next ball 45% of the time. This is a typical baseline for ball-by-ball prediction, as cricket is highly stochastic (predicting a 4 vs. a 6 or a 0 vs. a 1 is difficult without complex spatial data).


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Feature Importance
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names = X_train.columns

plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), feature_names[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
# plt.savefig('feature_importance.png')
```


2. Feature Importance The chart below shows which factors were most important for the model's decisions:

Key Insight: The most important features were innings_runs_so_far, total_balls_bowled, and innings_wickets_so_far. This suggests the state of the game (pressure of the score/wickets) influences the outcome more than just the specific batter or bowler names.

3. Classification Report

The model is good at predicting 0s (Dots) and 1s (Singles), which are the most common outcomes.

It struggles to predict 4s and 6s precisely, often confusing them with other outcomes due to class imbalance (there are fewer 6s than 1s).

# Run the All  Model and  find the best model 
 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.svm import LinearSVC, LinearSVR
from sklearn.linear_model import LogisticRegression, LinearRegression, SGDClassifier, SGDRegressor
from sklearn.naive_bayes import GaussianNB
```

```python
# Scale Data: Crucial for SVM, SGD, and Logistic Regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# --- DEFINE ALL MODELS ---
models = {
    # Classification Techniques
    "Random Forest Classifier": RandomForestClassifier(n_estimators=50, max_depth=10, n_jobs=-1, random_state=42),
    "Decision Tree Classifier": DecisionTreeClassifier(max_depth=10, random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, n_jobs=-1, random_state=42),
    "Naive Bayes": GaussianNB(),
    "SVM (Linear SVC)": LinearSVC(dual=False, random_state=42),
    "SGD Classifier (Gradient Descent)": SGDClassifier(loss='log_loss', max_iter=1000, n_jobs=-1, random_state=42),
    
    # Regression Techniques
    "Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(n_estimators=50, max_depth=10, n_jobs=-1, random_state=42),
    "Decision Tree Regressor": DecisionTreeRegressor(max_depth=10, random_state=42),
    "SVM (Linear SVR)": LinearSVR(random_state=42, max_iter=1000),
    "SGD Regressor (Gradient Descent)": SGDRegressor(max_iter=1000, random_state=42)
}
```

```python
# --- TRAIN & EVALUATE ---
print("Training Models...")
results = {}

for name, model in models.items():
    # Use scaled data for Linear/SVM/SGD models, raw data for Trees
    if name in ["Logistic Regression", "Naive Bayes", "SVM (Linear SVC)", "SGD Classifier (Gradient Descent)",
                "Linear Regression", "SVM (Linear SVR)", "SGD Regressor (Gradient Descent)"]:
        X_tr, X_te = X_train_scaled, X_test_scaled
    else:
        X_tr, X_te = X_train, X_test
        
    # Train
    model.fit(X_tr, y_train)
    
    # Predict
    y_pred = model.predict(X_te)
    
    # Evaluate: If regression, round prediction to nearest whole number
    if "Regressor" in name or "Linear Regression" in name or "SVR" in name:
        y_pred_class = np.round(y_pred).astype(int)
    else:
        y_pred_class = y_pred
        
    acc = accuracy_score(y_test, y_pred_class)
    results[name] = acc
    print(f"{name}: {acc:.4f}")
```

```python
# --- CONCLUSION ---
best_model = max(results, key=results.get)
print(f"\nBest Model: {best_model} with {results[best_model]:.2%} Accuracy")
```
